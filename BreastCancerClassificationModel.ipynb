**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split #Split dataset into training data and testing data
from sklearn.linear_model import LogisticRegression #Logistic Regression model
from sklearn.metrics import accuracy_score #Evaluation metric
```

#**Data Collection & Processing**

```python
#loading the data
BreastCancerDataset = sklearn.datasets.load_breast_cancer()
print(BreastCancerDataset)
```

```python
#Loading the data to a Data Frame
myDataFrame = pd.DataFrame( BreastCancerDataset.data, columns = BreastCancerDataset.feature_names)

#print first 5 rows of the data frame
myDataFrame.head()
```

```python
#adding the taget column to the data frame
myDataFrame ['Diagnosis'] = BreastCancerDataset.target
#print the last 5 rows of the data frame
myDataFrame.tail()
```

#**Train Test Split**

**Splitting input features(column: '*mean radius*'-'*worst fractal dimension*') and the target column (column: '*Diagnosis*')**

Set all the input features as x-axis (axis = 0) and the target column as y-axis(axis = 1)

```python
X = myDataFrame.drop(columns = 'Diagnosis', axis = 1)
Y = myDataFrame['Diagnosis']

print(X)
```

```python
print(Y)
```

**Split the data into Training Data & Testing Data:**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
#0.2 defines that 20% of the dataset will be testing data (rest are training data)
#train_test_split gives different data each time - random_state causes data to be split in the same manner

print(X.shape, X_train.shape, X_test.shape)
```

Note: No parenthesis is use after `.shape` as it returns a *tuple*-result

Note: Original data set contains *569* data-points, 80%(0.8) of the data comes into X_train(455) and 20%(0.2) of the data comes into X_test(114)

#**Model training (using Logistic Regression Model):**

```python
# Logistic Regressionn is efficient for binary classification
Model = LogisticRegression() #Loading an instance of the LR Model


#Training the LR Model using training data
Model.fit(X_train, Y_train) #Analyzes the relation between both arguments - (referring to the input features)
```

The Model now acknowledges how the dataset looks like when `Diagnosis` is 0 or 1 (Benign or Malignant)

#**Model Evaluation (Accuracy Score)**

For Classification problem we use a metric called ***Accuracy Score***

[*`from sklearn.metrics import accuracy_score #Evaluation metric`*]

```python
#Accuracy on Training data
XTrainPrediction = Model.predict(X_train) #Predict whether the `Diagnosis` is 0 or 1
TrainingDataAccuracy = accuracy_score(Y_train, XTrainPrediction)
print("Accuracy on Training Data = ", TrainingDataAccuracy)
```

Note:
*   Y_train is the true value and XTrainPrediction is the predicted value by the model
*  accuracy_score(parameter1, parameter2) compares the accuracy of two parameters
*  0.9472527472527472 means 94.7% of cases(data points of Training Data) - The correct value is predicted by model

NOTE: Accuracy for Test Data is important

```python
#Accuracy on Test data
XTestPrediction = Model.predict(X_test) #Predict whether the `Diagnosis` is 0 or 1
TestingDataAccuracy = accuracy_score(Y_test, XTestPrediction)
print("Accuracy on Testing Data = ", TestingDataAccuracy)
```

NOTE:

* 0.9298245614035088 means 92.9% of cases(data points of Testing Data) - The correct value is predicted by model
*   We find **Accuracy Score** for both Training Data and Testing Data because sometimes our model tends to overfit.


**Overfitting** occurs when the model cannot generalize and fits too closely to the training dataset instead. Overfitting happens due to several reasons, such as: The training data size is too small and does not contain enough data samples to accurately represent all possible input data values.

#**Building a Predictive System**

```python
#Create a tuple as input data
InputData = (13.08,15.71,85.63,520,0.1075,0.127,0.04568,0.0311,0.1967,0.06811,0.1852,0.7477,1.383,14.67,0.004097,0.01898,0.01698,0.00649,0.01678,0.002425,14.5,20.49,96.09,630.5,0.1312,0.2776,0.189,0.07283,0.3184,0.08183)


#Convert to NumPy Array for Pre-Processing before feeding into Logistic Regression Model
ConvertedNumPyInputData = np.asarray(InputData)

#Reshape the NumPy Array to inform the model we want results for only one datapoint
ReshapedInputData = ConvertedNumPyInputData.reshape(1, -1)

Prediction = Model.predict(ReshapedInputData)
print("Prediction:", Prediction)

if (Prediction[0] == 0):
  print('The Breast Cancer is Malignant')
else:
    print('The Breast Cancer is Benign')
```

NOTE:
The `-1` in `reshape(1, -1)` is a placeholder that tells NumPy to automatically determine the appropriate number of columns based on the length of the input array.

Explanation:

- **`reshape(1, -1)`**: The `1` means you want one row, and `-1` means that NumPy should figure out the number of columns automatically based on the size of the original array.
  
  For example, if your original array has 30 elements (like the tuple provided), the resulting reshaped array will have a shape of `(1, 30)`.

This is often done before feeding the data into a machine learning model, which typically expects input data to be in a 2D array format, where each row represents a single data point (sample) and each column represents a feature of that data point.
